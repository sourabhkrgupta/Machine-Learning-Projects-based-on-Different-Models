```python
# Importing basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Importing the dataset
dataset = datasets.load_iris()
```

```python
# Dataset Segregation
x= dataset.data
y= dataset.target
names = dataset.target_names
```

```python
# Finding the PCA clustering to the dataset with n=2
from sklearn.decomposition import PCA
model= PCA(n_components=2)
y_means = model.fit(x).transform(x)
```

```python
# variance Percentage
plt.figure()
colors = ['red','green','orange']

for color, i, target_name in zip(colors,[0,1,2],names):
    plt.scatter(y_means[y==i,0],y_means[y == i, 1], color=color, lw = 2, label=target_name)
plt.title('IRIS Clusterring')
plt.legend()
plt.show()
```

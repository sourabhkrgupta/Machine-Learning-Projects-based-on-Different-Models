```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('digit.csv')
data
```


### Segregate Dataset into X(Input/Independent variables) and Y(Output/Dependent variables)

```python
X= data.iloc[:,1:]
print(X.shape)
```

```python
Y = data.iloc[:,0]
print(Y.shape)
```

```python
# Splitting dataset into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of Model: {0}%".format(accuracy_score(y_test,y_pred)*100))
```

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(criterion='entropy')
model2 = RandomForestClassifier(criterion='entropy',n_estimators=90)
model3 = RandomForestClassifier(n_estimators=110)
model4 = RandomForestClassifier(criterion='log_loss')

model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
model3.fit(X_train,y_train)
model4.fit(X_train,y_train)

y_predmodel1 = model1.predict(X_test)
y_predmodel2 = model2.predict(X_test)
y_predmodel3 = model3.predict(X_test)
y_predmodel4 = model4.predict(X_test)

print("Accuracy of Model1: {0}%".format(accuracy_score(y_test,y_predmodel1)*100))
print("Accuracy of Model2: {0}%".format(accuracy_score(y_test,y_predmodel2)*100))
print("Accuracy of Model3: {0}%".format(accuracy_score(y_test,y_predmodel3)*100))
print("Accuracy of Model4: {0}%".format(accuracy_score(y_test,y_predmodel4)*100))
```
